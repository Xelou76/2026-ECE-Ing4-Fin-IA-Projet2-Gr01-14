```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
print("✅ Imports OK")
```

```python
df = pd.read_csv("../data/creditcard.csv")
print(f"Shape : {df.shape}")
print(f"Colonnes : {list(df.columns)}")
df.head()
```

```python
fraud = df[df['Class'] == 1]
normal = df[df['Class'] == 0]

print(f"Transactions normales : {len(normal)} ({len(normal)/len(df)*100:.2f}%)")
print(f"Transactions frauduleuses : {len(fraud)} ({len(fraud)/len(df)*100:.2f}%)")

plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df, palette=['green','red'])
plt.title("Distribution des classes (0=Normal, 1=Fraude)")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,4))

axes[0].hist(normal['Amount'], bins=50, color='green', alpha=0.7)
axes[0].set_title("Montants - Transactions normales")
axes[0].set_xlabel("Montant (€)")

axes[1].hist(fraud['Amount'], bins=50, color='red', alpha=0.7)
axes[1].set_title("Montants - Transactions frauduleuses")
axes[1].set_xlabel("Montant (€)")

plt.tight_layout()
plt.show()

print(f"Montant moyen fraude : {fraud['Amount'].mean():.2f}€")
print(f"Montant moyen normal : {normal['Amount'].mean():.2f}€")
```

```python
plt.figure(figsize=(16,10))
corr = df.corr()
sns.heatmap(corr, cmap='coolwarm', center=0, linewidths=0.5)
plt.title("Matrice de corrélation")
plt.show()
```
